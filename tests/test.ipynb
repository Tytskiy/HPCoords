```python
np.random.seed(42)
x = 2*np.pi*np.random.random(100)
sinx = np.sin(x)
cosx = np.cos(x)
```

```python
data = pd.DataFrame({"x": x, "sinx": sinx, "cosx": cosx})
data.head(5)
```

```python
hpcoords.parallel_coordinates(
    data,
    columns=["cosx", "x", "sinx"],
    directions=[1, -1, 1],
    spline=True,
    alpha=0.4,
    figsize=(14, 7),
    lw=1.5,
)
plt.show()
```
